```python
import pymysql
from sqlalchemy import create_engine
```

```python
def openConnection(user, password, host, database):
        try:
            sqlEngine = create_engine('mysql+pymysql://{}:{}@{}/{}'.format(user, password, host, database), pool_recycle=3600)
            databaseConnectionObject = sqlEngine.connect()

            return databaseConnectionObject
        
        except Exception as e:
            print("Could not establish connection to predictions database: ", e)
```

```python
def closeConnection(databaseConnectionObject):
    try:
        databaseConnectionObject.close()

    except Exception as e:
        print("Could not close database connection : ", e)
```

```python
def loadSQLScript(filepath):
    try:
        f = open(filepath, 'r')
        sqlFile = f.read()
        f.close()

        return sqlFile

    except Exception as e:
        print("Could not read .sql file: ", e)
```
