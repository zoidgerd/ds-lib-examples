```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
#
# https://docs.python-requests.org/en/latest/
#
def get_token(value1, value2):
    try:
        url = f'https://your.api.urp/api/params/here/token'

        payload = {
            "placeholder1": value1,
            "placeholder2": value2
        }
        
        response = requests.post(url, data=payload)
        response_json = json.loads(response.text)
        response = response_json.get('data_to_fetch')

        return response

    except Exception as e:
        print('Could not make request: ', e)
```
