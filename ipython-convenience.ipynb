### Introspection

```python
# object introspection:
a = 1
b = [1, 2, 3, 4]

# prefixing an object with a '?' gives you info about it
?a
```

```python
# you can get the source code of a function by prefixing it with ??
def add_numbers(a, b):
    c = a+b
    return c

??add_numbers
```

```python
# in large scripts you can search for functions using * for wildcard searches

# local functions
*add*?

# functions in packages e.g. numpy
np.*load*?
```
